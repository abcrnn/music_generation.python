## Music Generation: Data and Features
### Different music format
#### Midi format
Musical Instrument Digital Interface (MIDI) is a technical standard that describes a protocol, a digital interface and
connectors for interoperability between various electronic musical instruments, softwares and devices [2]

MIDI format carries event message that corresponding to real note performances. 2 most importance method is `note_on`, to indicate that a note is played and `note_off`, to indicate that note is ends. 
 
 An example of an excerpt from a midi file(turned in to ASCII) readable format. second column indicate the pitch, 3 columns indicate if note is on or off.

#### abc format
Abc notation are simple and powerful ASCII notation format. Abc is widely used to distribute tunes, particularly on the internet. A tune notated in abc can be directly read using multiple open source software like EasyAbc which can be converted to mp3 or music sheet.
Abc file contains a header section and tunes section. The header section is where the author can describe the music or make comments on the tunes and the tunes section is where music notes are placed. A complete example of a simple midi song in abc notation is shown in Figure 3. Example in Figure 3 shows an example including 4 header, where ‘X’ indicate the idx of the tunes, ‘T’ describe the title, ‘M’ describes the time signature of the tunes, and most importantly ‘K’ describes the key signature, the arrangement of sharp or flat signs on particular lines and space of musical notes.
